```python
import pandas as pd
```

```python
df_ratings = pd.read_csv('movie_data/ratings.csv')
df_metadata = pd.read_csv('movie_data/movies_metadata.csv')
```

```python
df_ratings
```

```python
df_metadata
```

```python
df_ratings = df_ratings.drop_duplicates()
df_metadata = df_metadata.drop_duplicates()
```

```python
df_metadata = df_metadata.drop(columns = ["budget", "homepage", "release_date", "revenue", "runtime", "status", "video"])
```

```python
df_metadata = df_metadata.drop(columns = "belongs_to_collection")
```

```python
df_ratings = df_ratings.fillna('')
df_metadata = df_metadata.fillna('')
```

```python
df_metadata = df_metadata[df_metadata['original_language'] == 'en']
```

```python
df_metadata
```

```python
df_ratings.to_csv('cleaned_ratings.csv', index=False) 
df_metadata.to_csv('cleaned_metadata.csv', index=False) 
```
